**import packages**

```python
import pandas as pd
import numpy as np
import xgboost as xgb
```

**note that label column in the labels file was concatenated to feature file**

```python
dfo = pd.read_excel('D:/Work&Study/Competetions/Vodafon/Tasks/Task1/feature.xlsx',sheet_name='feature')
df=dfo.iloc[1:515]
df
```

# Preprocess the data and transform objects to categorical data then apply hot one encoding transformation

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Unnamed: 132'].unique()
```

```python
pd.get_dummies(df,df.columns[df.dtypes == 'object'])
df.columns[df.dtypes == 'object']
```

```python
# Check For Unique Values of Each Column
for i in df.columns[df.dtypes == 'object']:
    print(i,"::",df[i].unique())
```

```python
df2 = df
for i in df.columns[df.dtypes == 'object']:
    print(i,"::")
    df2[i] = df[i].astype('category')
    df2[i] = df2[i].cat.codes
    
    
```

```python
df2.dtypes
```

```python
df2.shape
```

**split features from label column**

```python
X= df2.iloc[:,0:132]
```

```python
X.head()
```

```python
df2['Unnamed: 132'].unique()
```

```python
y=df2['Unnamed: 132']
```

**split data into training and test data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape
```

**prepare an instance from XGB classifier to be trained on the data**


```python
from sklearn.metrics import accuracy_score
```

```python
xgb_classifier = xgb.XGBClassifier()
```

# Transform data into matrix form to be prepared for XGB

```python
X_matrix = X.as_matrix()
y_matrix = y.as_matrix()
X_test_matrix=X_test.as_matrix()
xgb_classifier.fit(X_matrix,y_matrix)
```

```python
predictions = xgb_classifier.predict(X_test_matrix)
```

```python
predictions
```

```python
print("Accuracy of Model::",accuracy_score(y_test,predictions))
```

**Predict a single example**

```python
print("Prediction By XGBoost::",xgb_classifier.predict(X_test.iloc[1])) # double square braces if we don`t want to transform to matrix
```

**Show features weight and the probability per each example**

```python
import eli5
from eli5 import show_weights
show_weights(xgb_classifier)
```

```python
eli5.show_prediction(xgb_classifier, X_test_matrix[1], show_feature_values=True)
```
